```python
def summarize_casualty_type(casualty_type):
    """
    Summarizes the type of casualty involved in the accident.

    Parameters:
        casualty_type (str): The type of casualty.

    Returns:
        str: A summarized category representing the type of casualty.
    """
    if 'Car' in casualty_type:
        return 'Car Occupant'
    elif 'Motorcycle' in casualty_type:
        return 'Motorcycle Rider or Passenger'
    elif 'Goods' in casualty_type:
        return 'Goods Vehicle Occupant'
    elif 'Bus' in casualty_type or 'Minibus' in casualty_type:
        return 'Bus/Minibus Occupant'
    elif casualty_type == 'Pedestrian':
        return 'Pedestrian'
    elif casualty_type == 'Cyclist':
        return 'Cyclist'
    else:
        return 'Others'

# Apply the function to create a new column with summarized casualty types
df_cleaned['casualty_type_summary'] = df_cleaned['casualty_type'].apply(summarize_casualty_type)
```

We categorized the pedestrian's location at the time of the accident into broader categories.The categories are as follows:
- Unknown or other: Pedestrian location is unknown or falls into other unspecified categories.
- Pedestrian Crossing: Pedestrians involved in accidents while crossing at designated pedestrian crossings or zig-zag approaches.

```python
def summarize_pedestrian_location(pedestrian_location):
    if 'Unknown' in pedestrian_location:
        return 'Unknown or other'
    elif 'pedestrian crossing' in pedestrian_location or 'zig-zag' in pedestrian_location:
        return 'pedestrian crossing'
    elif 'carriageway' in pedestrian_location:
        return 'In carriageway'
    elif 'footway' in pedestrian_location:
        return 'On footway or verge'
    elif pedestrian_location == 'Not a Pedestrian':
        return pedestrian_location
    else:
        return 'Unknown or other'

df_cleaned['pedestrian_locatione_summary'] = df_cleaned['pedestrian_location'].apply(summarize_pedestrian_location)    
```

We categorized the movement of pedestrians involved in the accident into broader categories. The summarized categories, along with their respective counts, are illustrated in Figure 3. These categories include:
- Not a Pedestrian: Individuals categorized as non-pedestrians involved in accidents.
- Crossing: Pedestrians involved in accidents while crossing the road.
- Stationary in Carriageway: Pedestrians involved in accidents while standing or playing in the carriageway.
- Walking in Carriageway: Pedestrians involved in accidents while walking along the carriageway, either facing or back to traffic.
- Unknown or Other: Pedestrians involved in accidents with movement patterns not fitting into the above categories.


```python
# Define function to categorize pedestrian movement
def categorize_pedestrian_movement(movement):
    if 'Pedestrian' in movement:
        return 'Not a Pedestrian'
    elif 'Crossing' in movement:
        return 'Crossing'
    elif 'standing' in movement or 'playing' in movement:
        return 'Stationary in Carriageway'
    elif 'Walking' in movement:
        return 'Walking in Carriageway'
    else:
        return 'Unknown or Other'
```

```python
def plot_age_band_of_casualty(data: pd.DataFrame, my_palette: str, title: str = '') -> None:
    """
    Plot the count of occurrences for each age band of casualty.

    Parameters:
    - data (pd.DataFrame): DataFrame containing the data to be plotted.
    - my_palette (list): List of colors to be used for the plot.
    - title (str, optional): Title of the plot. Defaults to an empty string.

    Returns:
    - None
    """

    # Define the order of the bars
    age_band_order = ['0 - 5', '6 - 10', '11 - 15', '16 - 20', '21 - 25',
                      '26 - 35', '36 - 45', '46 - 55', '56 - 65', '66 - 75', 'Over 75']

    # Plot count of occurrences with specified order
    sns.countplot(data=data, x='age_band_of_casualty', order=age_band_order, palette=my_palette)

    # Set labels and title
    plt.xlabel('Age Band of Casualty')
    plt.ylabel('Count')
    plt.title(title)

    # Show plot
    plt.xticks(rotation=45)  # Rotate x-labels for better readability if needed
    plt.tight_layout()  # Adjust layout to prevent overlapping
    plt.show()
```

```python
def column_count_plot(df: pd.DataFrame, column_name: str, my_palette: list, 
                      hue_column: str = None, title: str = '') -> None:
    """
    Plot count of occurrences of each category in a categorical column.

    Parameters:
    - df (pd.DataFrame): DataFrame containing the data.
    - column_name (str): Name of the column to be summarized.
    - my_palette (list): List of colors to be used for the plot.
    - hue_column (str, optional): Name of the column to use for color encoding (if any). Defaults to None.
    - title (str, optional): Title of the plot. Defaults to an empty string.

    Returns:
    - None
    """

    # Plot count of occurrences with the darker palette
    ax = sns.countplot(data=df, x=column_name, hue=hue_column, palette=my_palette)

    # Set labels and title
    ax.set_xlabel(column_name.capitalize())
    ax.set_ylabel('Count')
    ax.set_title(title)

    # Rotate x-labels for better readability if needed
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

    # Annotate each bar with its count
    for p in ax.patches:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                    textcoords='offset points')

    # Show plot
    plt.show()
```

```python
def plot_donut_chart(df: pd.DataFrame, feature: str, title: str = '', my_palette: str = 'ch:s=-.2,r=.6') -> None:
    """
    Plots a donut chart based on the percentage of each category in a specified feature.

    Parameters:
    - df (pd.DataFrame): DataFrame containing the data.
    - feature (str): The column name representing the categorical feature.
    - title (str, optional): Title of the plot. Defaults to an empty string.
    - my_palette (str, optional): Palette for coloring the pie segments. Defaults to 'ch:s=-.2,r=.6'.

    Returns:
    - None
    """
    # Calculate percentage of each category in the feature
    category_counts = df[feature].value_counts(normalize=True) * 100

    # Set the color palette
    sns.set_palette(my_palette)

    # Plot
    plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)

    # Draw a circle at the center of the pie to make it a donut chart
    centre_circle = plt.Circle((0, 0), 0.80, fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)

    # Equal aspect ratio ensures that pie is drawn as a circle
    plt.axis('equal')

    # Add a title
    plt.title(title)
```

```python
def plot_categorical_distribution(df: pd.DataFrame, x_column: str, hue_column: str, title: str = '', 
                                  rotation: int = None) -> None:
    """
    Plots the distribution of a categorical variable based on two categorical variables.

    Parameters:
    - df (pd.DataFrame): DataFrame containing the data.
    - x_column (str): The column name representing one categorical variable.
    - hue_column (str): The column name representing another categorical variable for comparison.
    - title (str, optional): Title of the plot. Defaults to an empty string.
    - rotation (int, optional): Rotation angle for x-axis labels. Defaults to None.

    Returns:
    - None
    """
    # Group by the two categorical variables and calculate counts
    counts = df.groupby([x_column, hue_column]).size().reset_index(name='Count')

    # Calculate total count for each category in x_column
    total_counts = counts.groupby(x_column)['Count'].transform('sum')

    # Calculate percentages
    counts['Percentage'] = np.round((counts['Count'] / total_counts) * 100, 1)

    # Plot
    plt.figure(figsize=(15, 6))
    ax = sns.barplot(data=counts, x=x_column, y='Percentage', hue=hue_column, palette='ch:s=-.2,r=.6')

    # Set x-axis label
    plt.xlabel(x_column, fontsize=15)
    plt.xticks(fontsize=12, rotation=90)

    # Set y-axis label
    plt.ylabel('Percentage', fontsize=15)
    plt.yticks(fontsize=12)

    # Show the values on each bar
    for p in ax.patches:
        if f"{p.get_height()}%" != '0%':
            ax.annotate(f"{p.get_height()}%", (p.get_x() + p.get_width() / 2., p.get_height() + .5),
                        ha='center', va='center', fontsize=15, color='black', xytext=(0, 5),
                        textcoords='offset points', rotation=rotation)

    # Show the plot
    plt.title(title)
    plt.legend(title=hue_column, title_fontsize='13')
    plt.show()
```

```python
# peak 10000 data from each class as train data

train_data_size = 500

# Separate data by class
slight_data = df[df['casualty_severity_mixed'] == 0].sample(n=1100, random_state=42)
serious_data = df[df['casualty_severity_mixed'] == 1].sample(n=1000, random_state=42)


# Combine data from each class into training data
train_data = pd.concat([slight_data, serious_data])

# Remove training data from original data to create testing data
test_data = df.drop(train_data.index)

# Separate features and target variable
X_train = train_data.drop(['casualty_severity_mixed'], axis=1)
y_train = train_data['casualty_severity_mixed']

X_test = test_data.drop(['casualty_severity_mixed'], axis=1)
y_test = test_data['casualty_severity_mixed']
```

```python
def plot_confusion_matrix(y_true, y_pred):    
    # Calculate confusion matrix
    conf_matrix = confusion_matrix(y_true, y_pred)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g', 
                xticklabels=['slight', 'fatal and serious'], 
                yticklabels=['slight', 'fatal and serious'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, recall_score

# Instantiate the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42, n_estimators=100)

# Train the classifier on the training data
rf_classifier.fit(X_train, y_train)

# Predict on the training and test data
train_predictions = rf_classifier.predict(X_train)
test_predictions = rf_classifier.predict(X_test)

# Calculate recall for training set

train_recall = recall_score(y_train, train_predictions, average='weighted')

# Calculate recall for test set

test_recall = recall_score(y_test, test_predictions, average='weighted')

# Display the confusion matrices and recall scores
print("Confusion Matrix (Training Set):\n")
plot_confusion_matrix(y_train, train_predictions)
print("Recall (Training Set):", train_recall)
print("\nConfusion Matrix (Test Set):\n")
plot_confusion_matrix(y_test, test_predictions)
print("Recall (Test Set):", test_recall)
```
